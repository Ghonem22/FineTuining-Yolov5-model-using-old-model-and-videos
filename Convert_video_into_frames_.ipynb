```python
import yaml
import os
from os import listdir
from os.path import isfile, join
from pathlib import Path
import cv2
import glob
import shutil

def get_frames_margin(videos, total):
    total_frames = 0
    for v in videos:
        cap = cv2.VideoCapture(v)
        total_frames += int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        
    return total_frames // total

with open(r'utilities/configs.yml') as file:
    # The FullLoader parameter handles the conversion from YAML
    # scalar values to Python the dictionary format
    config = yaml.load(file, Loader=yaml.FullLoader)

if config['project_path']:
    os.chdir(config['project_path'])
  
videos = [f"{config['videos_folder_name']}/{f}" for f in listdir(config['videos_folder_name']) if isfile(join(config['videos_folder_name'], f))]
videos = [f for f  in videos if f[-3:] in config['videos_extensions']]

Path(config['saving_path']).mkdir(parents=True, exist_ok=True)

frames_margin = get_frames_margin(videos, config['frames_number'])
```
